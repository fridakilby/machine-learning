# Code

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.impute import SimpleImputer

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
```

```python
# Below, set your own path where you have stored the data file. 
housing = pd.read_csv(r'housing.csv')
```

## EDA

```python
print(housing.info()) # printing info of data in housing
```

# 2. Drop column ocean proximity

```python
housing = housing.drop(['ocean_proximity'], axis=1) # delete column with data-type object
```

```python
print(housing.info()) # printing info again to check that column with object is deleted
```

# 3. Split data into training set

```python
# split data to training and test-set, X is all columns except house value, Y is house value
X = housing.drop(["median_house_value"], axis=1)
y = housing["median_house_value"]

X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.2, random_state=21)
```

```python
print(X_train.info(), y_train.info()) # print info to see that the sets are ok
```

```python
# test to se if correlation matrix will give me info
import seaborn as sns
df_temp = X_train.copy()
df_temp['target'] = y_train
correlation_matrix = df_temp.corr()
sns.heatmap(correlation_matrix, annot=True)
# Heatmap dosen't show any big correlations what I can see
```

# 4. Replace missing values

```python
imputer_median = SimpleImputer(strategy="median") # calculate median using SimpleImputer
```

```python
X_train['total_bedrooms']=imputer_median.fit_transform(X_train[['total_bedrooms']])
X_test['total_bedrooms']=imputer_median.transform(X_test[['total_bedrooms']])
# Impter_median makes X_train_median to an nparray instead of dataframe, so I cant check with info if all data was replaced
```

```python
print(X_train, type(X_train))
```

```python
print(X_train.head())
```

```python
# printing some plots to see if I can see a pattern from the data, some patterns can be seen
from pandas.plotting import scatter_matrix
some_attributes = ["housing_median_age", "total_rooms", "total_bedrooms", "population"]
scatter_matrix(X_train[some_attributes], figsize=(12,8))
```

# 5. Linear regression and lasso model

```python
# calculation RMSE for linear regression of the train-set using cross validation and 5 folds
lin_reg = LinearRegression()
scores_lr = cross_validate(lin_reg, X_train, y_train, cv=5, scoring = 'neg_mean_squared_error')["test_score"]
print('RMSE for iteration:', np.sqrt(-scores_lr))
print('RMSE:', np.sqrt(np.mean(-scores_lr)))
```

```python
# calculation RMSE for lasso regression of the train-set, first testing differens alpha then using cross validation and 5 folds
lasso = Lasso()
hyper_param_lasso = {'alpha':(1, 5 ,10, 15, 20)} # done test with smaller and bigger numbers 
lasso_reg = GridSearchCV(lasso, hyper_param_lasso, cv=5)
lasso_reg.fit(X_train, y_train)

print(lasso_reg.best_params_)

scores_lasso = cross_validate(lasso_reg, X_train, y_train, cv=5, scoring = 'neg_mean_squared_error')["test_score"]
print('RMSE for iteration:', np.sqrt(-scores_lasso))
print('RMSE:', np.sqrt(np.mean(-scores_lasso)))
```

```python
# since lower RMSE is better Linear regression is the preferd modell for this data, train model with LR
lin_reg.fit(X_train, y_train)
```

# 6. Evaluate

```python
# printing mean for test-set, and also a box plot and datatype. Just for measurment to the prediction
print(np.mean(y_test))
y_test.plot.box()
print(type(y_test))
```

```python
# predict values for y using test-set of data and linear regression, also calculate RMSE and print RMSE
# set squared to False to get RMSE, if set to true you get MSE
y_test_predict_lr = lin_reg.predict(X_test)
RMSE_test_data  = mean_squared_error(y_test, y_test_predict_lr, squared = False)
print(RMSE_test_data)
```

```python
# Calculate differens of RMSE and mean of test-data
(RMSE_test_data)/(np.mean(y_test))
```
